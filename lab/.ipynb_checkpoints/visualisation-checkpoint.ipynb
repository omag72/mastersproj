```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pymysql

# MySQL connection parameters
mysql_params = {
    "host": "mysql",
    "user": "myuser",
    "password": "mypassword",
    "database": "mydatabase",
}

# MySQL query to select data
query = "SELECT * FROM sales_data"

# Create a connection to MySQL and fetch data into a DataFrame
try:
    conn = pymysql.connect(**mysql_params)
    df = pd.read_sql(query, con=conn)
except Exception as e:
    print(f"Error: {e}")
    # Handle the error or exit the script

# Display the first few rows of the DataFrame
df.head()

# Summary statistics
summary_stats = df.describe()

# Pair plot for numeric columns
sns.pairplot(df)
plt.show()

# Countplot of Quantity sold for each day of the week
plt.figure(figsize=(10, 6))
sns.countplot(x='day_of_week', data=df, palette='viridis')
plt.title('Quantity Sold by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Quantity Sold')
plt.show()

# Group by category and sum the total sales
category_sales = df.groupby('category')['total_price'].sum().reset_index()

# Sort the categories by total sales in descending order
category_sales = category_sales.sort_values(by='total_price', ascending=False)

# Bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='category', y='total_price', data=category_sales, palette='muted')
plt.title('Total Sales by Category (Descending Order)')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.show()


```
